## Time-series data

- mámě body v časovém pořadí, tj. jde o **časovou řadu**
- ideálně jsou body po sobě jdoucí po stejně rozložených časových bodech

```python
import pandas as pd

df = pd.read_csv('.\konzumace_cukr.csv')
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# roky budou ve formátu dat
df['Rok'] = pd.to_datetime(df['Rok'], format='%Y')

# zanes měření
ax.plot(df['Rok'], df['CR'])
ax.plot(df['Rok'], df['USA'])

# doplň popisky
ax.legend()
ax.set_title('Konzumace cukru ČR vs USA')
plt.ylabel('Gramy/os/den')
plt.xlabel('Rok')

# ukaž
plt.show()
```

## Kategorie

- porovnáváme hodnotu, frekvenci popř. jiné měření pro různé **kategorie** dat

```python
import matplotlib.pyplot as plt

# příprava dat
data = {'Raven': 10, 'Pilsner': 15, 'Birell': 5, 'Budvar': 1}
names = list(data.keys())
values = list(data.values())

# vyvtoř sub-grafy
fig, ax = plt.subplots(1, 3, figsize=(9, 3))

# dobré a špatné reprezentace dat
ax[0].bar(names, values) # sloupcový graf
ax[1].scatter(names, values) # bodový diagram
ax[2].plot(names, values) # spojnicový diagram

fig.suptitle('Reprezentace kategorií')
```

#### Co když nás zajímá **poměr** ?

```python
import matplotlib.pyplot as plt

# připrav data
data = {'Raven': 10, 'Pilsner': 15, 'Birell': 5, 'Budvar': 1}
names = list(data.keys())
values = list(data.values())

# vykresli koláčový graf
fig, ax = plt.subplots()
_ = ax.pie(values, labels=names, autopct='%1.1f%%')

# ukaž
plt.show()
```

## Spojitá měření

- data tvoří "spojitou" množinu čísel a my chceme vidět jejich **rozdělení (distribuci)**

```python
import numpy as np

# inicializuj generátor
np.random.seed(42)

total_numbers = 100000

# inicializace prázdného pole
random_numbers = np.empty(total_numbers)

# generuj náhodná čísla
for i in range(total_numbers):
    random_numbers[i] = np.random.random()

# vykresli histogram
_ = plt.hist(random_numbers)

# ukaž
plt.show()
```

## Vztahy mezi proměnnými

- zajímá nás **vztah** mezi proměnnými, data zobrazím jako body (dále mohu proložit křivkou, obarvit dle kategorií...)
- takto mohu zjistit nějakou formu **závislosti** (korelaci) ale i objevit v bodech **shluky**

### Shluky

```python
import matplotlib.pyplot as plt
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2] # potřebujeme jen první dva sloupce
y = iris.target

plt.figure(figsize=(8, 6))

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='black')
plt.xlabel('Délka okvětních plátků (cm)')
plt.ylabel('Šířka okvětních plátků (cm)')

plt.show()
```

### Pozor korelace vs kauzalita

- korelace znamená vzájemný vztah mezi dvěma procesy, tj. pokud se jedna veličina mění, mění se korelativně druhá
- tj, pokud prokážeme korelaci mezi dvěma procesy, je pravděpodobné, že na sobě závisejí, to však **neznamená**, že jeden z nich by musel být **příčinou** a druhý **následkem**
- viz [Wiki](https://cs.wikipedia.org/wiki/Korelace_neimplikuje_kauzalitu)

```python
import pandas as pd

df = pd.read_csv('.\deti_doba_zivota.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# zanes měření
plt.scatter(df['PocetDeti'], df['DobaZivota'])

# doplň popisky
plt.ylabel('Očekávaná střední doba života')
plt.xlabel('Počet dětí')

# ukaž
plt.show()
```

- chceme porovnat základní statistiky (rozptyl, medián, průměr, [kvartily](https://www.hackmath.net/cz/kalkulacka/vypocet-kvartilu-q1-q2-q3)), ale máme mnoho naměřených hodnot a více datasetů

### Boxplot
- neukáže nám distribuci v detailní formě
- možná detekce odlehlých pozorování, tj. **outlierů** , **šikmosti**
- snadné porovnání **více datasetů najednou** 

```python
# příprava generátoru a generování dat
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)

# připrav data pro formát vhodný k vykreslení   
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

fig, ax = plt.subplots(figsize=(9, 6))

# vytvoř boxplot
bp = ax.boxplot(data_to_plot, meanline=True, showmeans=True)

plt.show()
```
